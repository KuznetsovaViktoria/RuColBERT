```python
import json
from collections import OrderedDict

DRIVE_DATA = "/content/drive/MyDrive/colbert_ru/datasets"
DRIVE_RESULTS = "/content/drive/MyDrive/colbert_ru/results"
os.makedirs(DRIVE_RESULTS, exist_ok=True)
```

```python
# === 2a. Собрать корпус из всех уникальных документов (train + dev) ===
corpus_path = f"{DRIVE_DATA}/corpus.jsonl"
if not os.path.exists(corpus_path):
    unique_docs = OrderedDict()
    for ex in miracl_train + miracl_dev:
        for text in [ex.positive] + ex.negatives:
            if text and text not in unique_docs:
                doc_id = f"doc_{len(unique_docs)}"
                unique_docs[text] = doc_id

    with open(corpus_path, "w") as f:
        for text, doc_id in unique_docs.items():
            json.dump({"id": doc_id, "text": text}, f, ensure_ascii=False)
            f.write("\n")
    print(f"Corpus saved: {len(unique_docs)} unique docs → {corpus_path}")
else:
    print(f"Corpus already exists: {corpus_path}")
```

```python
# === 2b. Сохранить запросы и qrels из MIRACL dev ===
# Строим text→doc_id маппинг из корпуса
with open(corpus_path) as f:
    text_to_id = {}
    for line in f:
        row = json.loads(line)
        text_to_id[row["text"]] = row["id"]

queries_path = f"{DRIVE_DATA}/queries.jsonl"
qrels_path = f"{DRIVE_DATA}/qrels.tsv"

with open(queries_path, "w") as fq, open(qrels_path, "w") as fr:
    for i, ex in enumerate(miracl_dev):
        qid = f"q_{i}"
        json.dump({"id": qid, "text": ex.query}, fq, ensure_ascii=False)
        fq.write("\n")

        # positive
        pos_doc_id = text_to_id.get(ex.positive)
        if pos_doc_id:
            fr.write(f"{qid}\t0\t{pos_doc_id}\t1\n")

        # negatives
        for neg_text in ex.negatives:
            neg_doc_id = text_to_id.get(neg_text)
            if neg_doc_id:
                fr.write(f"{qid}\t0\t{neg_doc_id}\t0\n")
```

```python
print(f"Queries: {queries_path} ({len(miracl_dev)} queries)")
print(f"Qrels:   {qrels_path}")
```

```python
import sys
sys.path.insert(0, "/content/colbert_ru_drive")

from index_and_retrieve import BM25Baseline
import json, time

# Загружаем корпус
with open(f"{DRIVE_DATA}/corpus.jsonl") as f:
    corpus = [json.loads(line) for line in f]

# Загружаем запросы
with open(f"{DRIVE_DATA}/queries.jsonl") as f:
    queries = [json.loads(line) for line in f]

print(f"Corpus: {len(corpus)} docs, Queries: {len(queries)}")
```

```python
# Строим BM25
t0 = time.time()
bm25 = BM25Baseline(
    corpus_texts=[d["text"] for d in corpus],
    doc_ids=[d["id"] for d in corpus],
)
print(f"BM25 index built in {time.time()-t0:.1f}s")

# Retrieval
t0 = time.time()
bm25_results = {}
for q in queries:
    hits = bm25.retrieve(q["text"], top_k=100)
    bm25_results[q["id"]] = hits
print(f"BM25 retrieval done in {time.time()-t0:.1f}s")
```

```python
# Сохраняем результаты
bm25_path = f"{DRIVE_RESULTS}/bm25_results.json"
with open(bm25_path, "w") as f:
    json.dump({"BM25": bm25_results}, f, ensure_ascii=False)
print(f"BM25 results -> {bm25_path}")
```

```python
import sys, json, time, torch
import safetensors.torch
from huggingface_hub import hf_hub_download

sys.path.insert(0, "/content/colbert_ru_drive/code/")

from modeling import build_model_and_tokenizer
from index_and_retrieve import ColBERTIndexer, ColBERTRetriever
from config import ModelConfig, IndexConfig

ZS_CHECKPOINT = "antoinelouis/colbert-xm"
COLBERT_XM_FAITHFUL = True
ENCODER_FINETUNE_REF = "xlm-roberta-base"
XMOD_ENCODER = "facebook/xmod-base"
XMOD_DEFAULT_LANG = "ru_RU"

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")


def load_colbert_xm_weights(model, hub_id=ZS_CHECKPOINT):
    path = hf_hub_download(hub_id, "model.safetensors")
    hf_sd = safetensors.torch.load_file(path)
    new_sd = {}
    for k, v in hf_sd.items():
        if k.startswith("roberta."):
            new_sd[k.replace("roberta.", "encoder.backbone.", 1)] = v
        elif k == "linear.weight":
            new_sd["encoder.linear.weight"] = v
        else:
            new_sd[k] = v
    inc = model.load_state_dict(new_sd, strict=False)
    print(
        f"ColBERT-XM load: missing_keys={len(inc.missing_keys)} "
        f"unexpected_keys={len(inc.unexpected_keys)}"
    )
    bb = model.encoder.backbone
    if hasattr(bb, "set_default_language"):
        bb.set_default_language(XMOD_DEFAULT_LANG)


encoder_name_zs = XMOD_ENCODER if COLBERT_XM_FAITHFUL else ENCODER_FINETUNE_REF
print(
    f"Zero-shot encoder: {encoder_name_zs} (faithful={COLBERT_XM_FAITHFUL}; "
    f"ref. для своей модели: {ENCODER_FINETUNE_REF})"
)

model_cfg = ModelConfig(encoder_name=encoder_name_zs, embedding_dim=128)
model, tokenizer = build_model_and_tokenizer(model_cfg)
if COLBERT_XM_FAITHFUL:
    load_colbert_xm_weights(model, ZS_CHECKPOINT)
model.eval()

with open(f"{DRIVE_DATA}/corpus.jsonl") as f:
    corpus = [json.loads(line) for line in f]

index_cfg = IndexConfig(
    index_dir="/content/drive/MyDrive/colbert_ru/index_zeroshot_colbert_xm_xmod",
    batch_size=256,
    save_fp16=True,
)
indexer = ColBERTIndexer(model, tokenizer, index_cfg, model_cfg)
indexer.build(
    texts=[d["text"] for d in corpus],
    doc_ids=[d["id"] for d in corpus],
)
```

```python
retriever = ColBERTRetriever(
    model, tokenizer,
    index_dir=index_cfg.index_dir,
    model_cfg=model_cfg,
    device=device,
)

with open(f"{DRIVE_DATA}/queries.jsonl") as f:
    queries = [json.loads(line) for line in f]

from tqdm import tqdm

zs_results = {}
for q in tqdm(queries, desc="Retrieving (zero-shot XMOD)"):
    hits = retriever.retrieve(q["text"], top_k=100)
    zs_results[q["id"]] = hits

zs_path = f"{DRIVE_RESULTS}/zeroshot_colbert_xm_faithful.json"
with open(zs_path, "w") as f:
    json.dump({"ColBERT-XM_zero-shot_xmod_faithful_ru_RU": zs_results}, f, ensure_ascii=False)
print(f"Zero-shot results -> {zs_path}")
```
